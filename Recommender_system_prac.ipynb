```python
import pandas as pd
```

```python
movie_data = pd.read_csv(r'movie.csv')
movie_data.head()
```

```python
len(movie_data.userId.unique())    # 4081
len(movie_data.movie.unique())     # 10
```

```python
# to make pivot table for better understanding :
movie_pivot_table = movie_data.pivot(index='userId', columns='movie', values='rating')
movie_pivot_table.head(25)
```

```python
movie_pivot_table.index = movie_data.userId.unique()
movie_pivot_table
```

```python
# replace null value with 0 :
movie_pivot_table.fillna(0, inplace=True)
movie_pivot_table
```

```python
# calculate cosine similarity between users :
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
users_simi = 1-pairwise_distances(movie_pivot_table, metric='cosine')
# to change diagonal value :
import numpy as np
np.fill_diagonal(users_simi, 0)
users_simi
```

```python
# similarity metrix shows similarity between each users :
users_simi_df = pd.DataFrame(users_simi)
users_simi_df
```

```python
# to change index and columns :
users_simi_df.index = movie_data.userId.unique()
users_simi_df.columns = movie_data.userId.unique()
users_simi_df
```

```python
# to find max similarity between users :
users_simi_df.idxmax(axis=1)
```

```python
# take out individual user data which has high similarity, (in our case we're taking user with Id 6 & 60.)
user_1 = movie_data[movie_data.userId == 2]
user_2 = movie_data[movie_data.userId == 8]
user_1                                          # both user has given rating to grumpier old man movie :
user_2
```

```python
# merge both users data for better recommendation :
pd.merge(user_1, user_2, on='movie', how='outer')
# from above code's output we can say that we need to recommend user_1 movies [toy story, heat, goldeneye]
```
