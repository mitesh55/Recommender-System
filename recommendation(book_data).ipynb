```python
import pandas as pd
```

```python
data = pd.read_csv(r"bookrec.csv", error_bad_lines=False, encoding="latin-1")
data.head()
# data.shape                # (10000, 4)
```

```python
# df = pd.get_dummies(data)
# df.head()
df.shape                 # (10000, 9662)
df = data.rename({"User.ID":"userid", "Book.Title":"booktitle", "Book.Rating":"rating"}, axis=1)
df.head()
```

```python
len(df.userid.unique())              # 2182
len(df.booktitle.unique())           # 9659
df.head()
```

```python
pivot_table = df.reset_index().pivot_table(index="userid", columns="booktitle", values="rating")
pivot_table                            # (2182, 9659)
```

```python
pivot_table.index = df.userid.unique()
pivot_table
```

```python
pivot_table.fillna(0, inplace=True)
pivot_table
```

```python
# calculate similarity between users :
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
import numpy as np

user_simi = 1-pairwise_distances(pivot_table, metric="cosine")
user_simi
np.fill_diagonal(user_simi, 0)
user_simi
```

```python
user_simi_df = pd.DataFrame(user_simi)
user_simi_df.head()
# user_simi_df.shape                         # (2182, 2182)
user_simi_df
```

```python
# change index and column :
user_simi_df.index = df.userid.unique()
user_simi_df.columns = df.userid.unique()
user_simi_df.head()
user_simi_df
```

```python
# to find max similarity between users :
user_simi_df.idxmax(axis=1)
```
